Question 1:- A digital payment platform recorded the following transaction amounts made by a customer during a single day.

transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]

Write a Python program that performs the following analysis:
1. Identify and display all transaction values that appear more than once.
2. Detect transactions greater than ₹5000 as these may require fraud monitoring.
3. Calculate the total spending for the day.
4. Create a new list containing only unique transaction values.
5. Sort the unique transaction list in descending order to view the largest transactions first.

```python
transactions = [1200, 3000, 250, 7800, 540, 3000, 9100, 250, 1500, 7800]
transaction_appear_once = []
transactions_appear_more_than_once = []
transactions_greater_than_5000 = []
total_spending = 0
transaction_unique_only = []
transaction_unique_descending_order = []

for t in transactions:
    if t in transaction_appear_once:
        transactions_appear_more_than_once.append(t)
    else:
        transaction_appear_once.append(t)

    if t > 5000:
        transactions_greater_than_5000.append(t)

    total_spending = total_spending + t

    if t not in transaction_unique_only:
        transaction_unique_only.append(t)



print(f"All transaction values appearing more than once: {transactions_appear_more_than_once}")
print(f"Transactions greater than ₹5000: {transactions_greater_than_5000}")
print(f"Total spending for the day: {total_spending}")
print(f"List containing only unique transaction values: {transaction_unique_only}")
transaction_unique_only.sort(reverse = True)
print(f"Unique transaction list in descending order: {transaction_unique_only}")
```

Question 2:- A dataset exported from a database contains the following customer age values.

ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]
Some entries are invalid because they are either negative numbers or missing values (None).

Write a Python program that:
1. Iterates through the dataset and removes all invalid age values.
2. Stores only valid ages in a new list.
3. Calculates the average age of valid customers.
4. Displays the list of customers whose age is greater than 30.
5. Prints how many invalid entries were removed during the cleaning process.

```python
ages = [25, 42, -1, 34, None, 52, 17, -5, 29, None, 46]
valid_ages = []
average_age = 0
ages_greater_than_30 = []
number_of_invalid_values = 0

for age in ages:
    if age is None or age <= 0:
        number_of_invalid_values = number_of_invalid_values + 1
    else:
        valid_ages.append(age)
        average_age = average_age + age
        if age > 30:
            ages_greater_than_30.append(age)

print(f"List of valid ages: {valid_ages}")
print(f"Average age of valid customers: {average_age/len(valid_ages)}")
print(f"List of customers whose age is greater than 30: {ages_greater_than_30}")
print(f"Number of invalid entries removed during cleaning process: {number_of_invalid_values}")
```

Question 3:- An analyst wants to compare the price of products listed on two e-commerce platforms.

The following data is provided:
products = ["Laptop","Phone","Tablet","Monitor","Headphones"]
amazon_prices = [75000, 45000, 30000, 20000, 5000]
flipkart_prices = [72000, 47000, 31000, 19500, 4800]

Write a Python program that:
1. Combines the product names and their prices into a single data structure representing a table.
2. Compares the price of each product across both platforms.
3. Determines which platform offers the lower price.
4. Calculates the difference between the two prices.
5. Displays the product name, cheaper platform, and price difference.

```python
products = ["Laptop","Phone","Tablet","Monitor","Headphones"]
amazon_prices = [75000, 45000, 30000, 20000, 5000]
flipkart_prices = [72000, 47000, 31000, 19500, 4800]
combineData = []

#Two columns have been added in zip to maintain platform name with lower price and price difference
for combine in zip(products, amazon_prices, flipkart_prices, range(len(products)), range(len(products))):
    combine = list(combine)
    if combine[1] > combine[2]:
        combine[3] = "Flipkart"
        combine[4] = combine[1] - combine[2]
    elif combine[1] < combine[2]:
        combine[3] = "Amazon"
        combine[4] = combine[2] - combine[1]
    else:
        combine[3] = "Amazon/Flipkart" 
        combine[4] = 0

    combineData.append(combine)

for product in combineData:
    print(f"Product Name: {product[0]}\t Cheaper Platform: {product[3]}\t Price Difference: {product[4]}")

```

Question 4:- A web platform records the hours when users log into the system.

login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
Write a Python program that:

1. Separates login hours into even-hour logins and odd-hour logins.
2. Identifies login hours that fall into peak activity time (hours greater than 5).
3. Sorts the login hours in ascending order.
4. Calculates how many logins occurred during peak hours.

```python
login_hours = [1, 3, 5, 7, 2, 6, 8, 4, 9, 10]
even_hours = []
odd_hours = []
login_hours_in_peak = []
number_of_login_in_peak = 0
for hour in login_hours:
    if hour%2 == 0:
        even_hours.append(hour)
    else:
        odd_hours.append(hour)

    if hour > 5:
        login_hours_in_peak.append(hour)
        number_of_login_in_peak = number_of_login_in_peak + 1

print(f"Even-hour logins: {even_hours}")
print(f"Odd-hour logins: {odd_hours}")
print(f"Login hour that fall into peak activity time (hours greater than 5): {login_hours_in_peak}")
login_hours.sort()
print(f"Login hours in ascending order: {login_hours}")
print(f"Number of logins occurred during peak hours: {number_of_login_in_peak}")
```

Question 5:- An e-commerce company exports its daily order log in the following format.
Each order contains multiple pieces of information:

orders = [
["ORD001","Laptop",75000,"Completed"],
["ORD002","Mouse",500,"Completed"],
["ORD003","Laptop",75000,"Returned"],
["ORD004","Keyboard",1500,"Completed"],
["ORD005","Laptop",75000,"Completed"],
["ORD006","Monitor",20000,"Fraud"],
["ORD007","Mouse",500,"Completed"],
["ORD008","Laptop",75000,"Completed"]
]

Each row represents an order with the following structure:

[Order_ID , Product_Name , Price , Status]

Where:
Order_ID → unique order identifier
Product_Name → product purchased
Price → price of the product
Status → order status (Completed, Returned, Fraud)

Your Task:

Write a Python program to simulate a simple order analysis system using loops, nested loops, conditional statements, and list operations.

Your program must perform the following steps.

1. Display All Orders
Using a nested loop, print the entire order table row by row so that each order’s details are displayed clearly.

2. Detect Suspicious Activity
The system suspects fraud when:
the same product appears more than 3 times, OR
an order has status "Fraud"

Write a program that scans the dataset and identifies suspicious products or orders.
You must use nested loops to compare product occurrences.

3. Calculate Total Revenue
Only Completed orders contribute to revenue.
Write code that:
iterates through the dataset
adds the price of all completed orders
prints the total revenue generated for the day.

4. Count Order Types
Using loops and conditional statements, count how many orders fall into each category:

Completed
Returned
Fraud
Store these counts in separate variables.

5. Generate a List of Unique Products
Create a new list containing unique product names present in the dataset.
You must implement this without using the set() function.

6. Product Sales Analysis
For each unique product, determine:
how many times it was ordered
total revenue generated by that product
This requires nested loops.

7. Identify the Most Sold Product
Based on the frequency of orders, determine which product appears the most in the dataset.

8. Remove Fraud Orders
Create a cleaned dataset by removing orders whose status is "Fraud".
Use list methods to store the cleaned data in a new list.

```python
orders = [ 
    ["ORD001","Laptop",75000,"Completed"], 
    ["ORD002","Mouse",500,"Completed"], 
    ["ORD003","Laptop",75000,"Returned"], 
    ["ORD004","Keyboard",1500,"Completed"], 
    ["ORD005","Laptop",75000,"Completed"], 
    ["ORD006","Monitor",20000,"Fraud"], 
    ["ORD007","Mouse",500,"Completed"], 
    ["ORD008","Laptop",75000,"Completed"] 
]

#1. Display All Orders
print("1. Display All Orders")
print(f"ID\t Name\t Price\t Status")
total_orders = len(orders)
i = 0
while i < total_orders:
    columns = len(orders[i])
    j = 0
    row = ""
    while j < columns:
        row = row + str(orders[i][j]) + "\t "
        j = j + 1
    print(row)
    i = i + 1
print("\n")

#2. Detect Suspicious Activity
print("2. Detect Suspicious Activity")
total_orders = len(orders)
product_scanned = []
suspicious_products = []
i = 0
while i < total_orders:
    if orders[i][1] not in product_scanned:
        j = 0
        appearance = 0
        while j < total_orders:
            if orders[i][1] == orders[j][1]:
                appearance = appearance + 1
            j = j + 1
            
        if appearance > 3 or orders[i][3] == "Fraud":
            suspicious_products.append(orders[i][1])
        product_scanned.append(orders[i][1])
    i = i + 1

print(f"A product appears more than 3 times, OR an order has status 'Fraud': {suspicious_products}\n")

#3. Calculate Total Revenue
print("3. Calculate Total Revenue")
total_revenue = 0
for order in orders:
    if order[3] == "Completed":
        total_revenue = total_revenue + order[2]

print(f"Total Revenue: {total_revenue}\n")

#4. Count Order Types
print("4. Count Order Types")
no_of_completed_orders = 0
no_of_returned_orders = 0
no_of_fraud_orders = 0
for order in orders:
    if order[3].lower() == "completed":
        no_of_completed_orders += 1
    if order[3].lower() == "returned":
        no_of_returned_orders += 1
    if order[3].lower() == "fraud":
        no_of_fraud_orders += 1

print(f"Number of Completed orders: {no_of_completed_orders}")
print(f"Number of Returned orders: {no_of_returned_orders}")
print(f"Number of Fraud orders: {no_of_fraud_orders}\n")

#5. Generate a List of Unique Products
print("5. Generate a List of Unique Products")
unique_products = []
for order in orders:
    if order[1] not in unique_products:
        unique_products.append(order[1])

print(f"List of Unique Products: {unique_products}\n")

#6. Product Sales Analysis
print("6. Product Sales Analysis")
most_ordered_product_no = 0
most_ordered_product_name = ""
for product in unique_products:
    no_of_orders = 0
    total_revenue = 0
    for order in orders:
        if order[1] == product:
            no_of_orders += 1
            total_revenue += order[2]
    if most_ordered_product_no < no_of_orders:
        most_ordered_product_no = no_of_orders
        most_ordered_product_name = order[1]
    print(f"Product Name: {product}\t No of Orders: {no_of_orders}\t Total Revenue: {total_revenue}")

#7. Identify the Most Sold Product
print("\n7. Identify the Most Sold Product")
print(f"Product appears the most in the dataset: {most_ordered_product_name}\n")

#8. Remove Fraud Orders
print("8. Remove Fraud Orders")
clean_dataset = list()
for order in orders:
    if order[3].lower() != "fraud":
        clean_dataset.append(order)

print(f"{clean_dataset}")
        
        
```
