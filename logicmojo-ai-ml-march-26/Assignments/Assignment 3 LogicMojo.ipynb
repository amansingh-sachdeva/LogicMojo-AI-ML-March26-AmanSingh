Q6. Handling Missing Data (Critical ML Step)

Context: Models cannot handle null values.

Dataset: Titanic

Tasks:

1. Fill missing age using median
2. Fill missing embarked using mode
3. Drop deck column
4. Explain:
Why different strategies are used for different columns

```python
import numpy as np
import pandas as pd
import seaborn as sns

df = sns.load_dataset("titanic")

#Fill missing age using median
df["age"] = df["age"].fillna(df["age"].median())

#Fill missing embarked using mode
df["embarked"] = df["embarked"].fillna(df["embarked"].mode()[0])

#Drop deck column
df = df.drop(columns=["deck"])

print(df)

#Explain: Why different strategies are used for different columns
print("Age => Missing values are filled with the median because: Age is continuous, so imputing with the most typical central value preserves distribution.")
print("Embarked => Missing values are filled with the mode because: For categorical features, you can't take an average, so the most common category will be a good replacement.")
print("Deck => Dropped because: Too many missing entries make imputation unreliable.")
```

Q13. Custom Feature Transformation using apply()

Context: Real ML requires custom logic transformation.

Dataset: Titanic

Tasks:

1. Create age_group:
Child / Adult / Senior
2. Use .apply()
3. Compute survival rate per group
4. Interpret:
Which segment has highest survival likelihood

```python
import numpy as np
import pandas as pd
import seaborn as sns

df = sns.load_dataset("titanic")

def age_group(age):
    if age < 18:
        return 'Child'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'

df["age_group"] = df['age'].apply(age_group)

survival_rates = df.groupby('age_group')['survived'].mean()
print(survival_rates)

print("""
Children often had the highest survival chances. Senior had lowest survival rate.
""")
```

Q14. Error Analysis Simulation (Model Evaluation Thinking)

Context: Understanding prediction errors is key in ML.

Tasks:

1. Generate 1000 random error values using NumPy
2. Compute:
Mean
Standard deviation
3. Identify outliers:
Using mean ± 2*std
4. Explain:
What high variance means in model performance

```python
import numpy as np

# 1. Generate 1000 random error values using NumPy
np.random.seed(42)
np_errors = np.random.normal(size=1000)

# 2. Compute mean and standard deviation
mean_error = np.mean(np_errors)
std_error = np.std(np_errors)

print("Mean:", mean_error)
print("Standard deviation:", std_error)

# 3. Identify outliers using mean ± 2*std
lower_bound = mean_error - 2 * std_error
upper_bound = mean_error + 2 * std_error
outliers = np_errors[(np_errors < lower_bound) | (np_errors > upper_bound)]
print("outliers:\n",outliers)
```
